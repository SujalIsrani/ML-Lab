```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Number of data points
n = 100

# Generate X between 0 and 10
X = np.random.rand(n) * 10

# True parameters
theta_true = [5, 2]  # θ0 = 5, θ1 = 2

# Generate Y with some noise
noise = np.random.randn(n)
y = theta_true[0] + theta_true[1] * X + noise

# Visualize
plt.scatter(X, y)
plt.title("Generated Data")
plt.xlabel("X")
plt.ylabel("y")
plt.show()
```

```python
def hypothesis(X, theta):
    return theta[0] + theta[1] * X
```

```python
def cost_function(X, y, theta):
    m = len(y)
    predictions = hypothesis(X, theta)
    return (1 / (2 * m)) * np.sum((predictions - y) ** 2)
```

```python
def compute_gradient(X, y, theta):
    m = len(y)
    predictions = hypothesis(X, theta)
    error = predictions - y
    d_theta0 = np.sum(error) / m
    d_theta1 = np.sum(error * X) / m
    return np.array([d_theta0, d_theta1])
```

```python
theta = np.zeros(2)
alpha = 0.01
iterations = 1000
cost_history = []

for i in range(iterations):
    grad = compute_gradient(X, y, theta)
    theta -= alpha * grad
    cost = cost_function(X, y, theta)
    cost_history.append(cost)

print("Final Parameters:", theta)
```

```python
plt.plot(range(iterations), cost_history)
plt.title("Cost Function Over Iterations")
plt.xlabel("Iterations")
plt.ylabel("Cost (MSE)")
plt.show()
```

```python
plt.scatter(X, y, label="Data")
plt.plot(X, hypothesis(X, theta), color='red', label="Fitted Line")
plt.title("Linear Regression Fit")
plt.xlabel("X")
plt.ylabel("y")
plt.legend()
plt.show()
```
