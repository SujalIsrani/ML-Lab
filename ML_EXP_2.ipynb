```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x):
    return x[0]**2 + 2 * x[1]**2
```

```python
def grad_f(x):
    df_dx1 = 2 * x[0]
    df_dx2 = 4 * x[1]
    return np.array([df_dx1, df_dx2])
```

```python
# Parameters
alpha = 0.1  # Learning rate
iterations = 50
x = np.array([4.0, 4.0])  # Initial guess

# plotting
path = [x.copy()]

for _ in range(iterations):
    gradient = grad_f(x)
    x = x - alpha * gradient
    path.append(x.copy())

print("Final x1, x2:", x)
print("Final function value f(x):", f(x))
```

```python
# grid
x1_vals = np.linspace(-5, 5, 100)
x2_vals = np.linspace(-5, 5, 100)
X1, X2 = np.meshgrid(x1_vals, x2_vals)
Z = X1**2 + 2 * X2**2

path = np.array(path)

# Plot
plt.figure(figsize=(8, 6))
plt.contour(X1, X2, Z, levels=30)
plt.plot(path[:, 0], path[:, 1], 'ro--', label='Gradient Descent Path')
plt.scatter(path[0, 0], path[0, 1], color='green', label='Start')
plt.scatter(path[-1, 0], path[-1, 1], color='black', label='End')
plt.title("Gradient Descent Optimization Path")
plt.xlabel("x1")
plt.ylabel("x2")
plt.legend()
plt.grid(True)
plt.show()
```
